```python
with open("day2.txt") as f:
    content = f.readlines()
content = [x.strip() for x in content] 
content
```

```python
checksum=[0,0]
for q in content:
    my_dict = {}
    for x in q:
        if x in my_dict:
            my_dict[x] += 1
        else:
            my_dict[x] = 1
    if 2 in my_dict.values():
        checksum[0] += 1
    if 3 in my_dict.values():
        checksum[1] += 1
checksum
```

```python
checksum[0] * checksum[1]
```

```python
import itertools
for a, b in itertools.combinations(content, 2):
    ok = 0
    for x, y in zip(a,b):
        if x != y:
            ok +=1
        if ok > 1:
            break
    if ok <= 1:
        print(a,b)
        break

out = ""
for s, t in zip(a,b):
    if s == t:
        out += s
print(out)
```
